```python
# Installs required packages if not already installed
!{sys.executable} -m pip install opencv-python
!{sys.executable} -m pip install os-sys

# IMPORTS
import cv2
import os, sys
from os.path import isfile, join
      
# Opens the Video file
cap= cv2.VideoCapture('vid/test-vid.avi')
i=0

# Seperates the file into individual frames
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == False:
        break
    cv2.imwrite('img/'+str(i)+'.jpg',frame)
    i+=1
 
cap.release()
cv2.destroyAllWindows()

n = 0

# Loops through all frames applies the filtering and outputs proccessed frames
while i != 0:
    
    # Assign input path
    path = 'img/'+str(n)+'.jpg'
    
    # Read input image
    image_in = cv2.imread(path)  
    
    # Convert image to greyscale
    image_grey = cv2.cvtColor(image_in, cv2.COLOR_BGR2GRAY)
    
    # Convert image to 2-bit black and white
    (thresh, image_bw) = cv2.threshold(image_grey, 245,255, cv2.THRESH_BINARY)
    
    # Remove any excessive noise from 2-bit conversion
    final = cv2.fastNlMeansDenoising(image_bw,40,10,7,21)
    
    # Write the processed frame into a file
    cv2.imwrite('output/'+str(n)+'.jpg', final)
    
    # Increment / decrement loop counters 
    n+=1
    i-=1

# Define input / output paths for video creation
pathIn= 'output/'
pathOut = 'final/video.mp4'

# Create an array to store all frames from input path
frame_array = []
files = [f for f in os.listdir(pathIn) if isfile(join(pathIn, f))]

# Sorting the frames into the correct order
files.sort(key = lambda x: int(x[:-4]))

# Loop through all processed frames and insert said frames into the array
for i in range(len(files)):
    filename=pathIn + files[i]
    
    # Read frame instance
    img = cv2.imread(filename)

    # Inserting the frame instance into an image array
    frame_array.append(img)
    
# Use populated array to write to a video file 
out = cv2.VideoWriter(pathOut,cv2.VideoWriter_fourcc(*'DIVX'), 24.0, (1296,972))

for i in range(len(frame_array)):
    # Writing to a image array
    out.write(frame_array[i])
    
out.release()
```
